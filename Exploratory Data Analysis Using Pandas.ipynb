```python
sns.heatmap(df.corr(numeric_only=True), annot = True)

plt.rcParams['figure.figsize'] = (12,4)
plt.show()
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df_continent_mean = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True)
df_continent_mean
```

```python
df_continent_mean = df_continent_mean.transpose()
df_continent_mean
```

```python
df_continent_mean.plot(marker='o', linestyle='-')

plt.title('Mean Ratings by Continent', fontsize=16, fontweight='bold')
plt.xlabel('Continent', fontsize=12)
plt.ylabel('Mean Rating', fontsize=12)
plt.grid(True, linestyle='-', alpha=0.9)
plt.legend(title='Ratings')
plt.tight_layout()
plt.show()
```
